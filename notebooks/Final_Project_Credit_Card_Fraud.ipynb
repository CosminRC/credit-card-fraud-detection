```python
plt.figure(figsize=(14,6))
sns.countplot(x='HourOfDay', hue='Class', data=df)
plt.title("Distribuția tranzacțiilor pe ore (normal vs fraudă)")
plt.xlabel("Ora zilei")
plt.ylabel("Număr tranzacții")
plt.xticks(range(24))
plt.legend(labels=['Normal', 'Fraud'])
plt.show()
```

„Graficul prezintă distribuția tranzacțiilor pe parcursul zilei, diferențiind între tranzacții normale și frauduloase. Se observă că majoritatea tranzacțiilor sunt non-fraudă, reflectând dezechilibrul datasetului. Totodată, fraudele apar pe tot parcursul zilei, cu ușoare variații ale frecvenței în funcție de oră, ceea ce sugerează că momentul tranzacției poate influența comportamentul fraudulos.

```python
fraud_rate_by_hour = df.groupby('HourOfDay')['Class'].mean() * 100
plt.figure(figsize=(14,6))
fraud_rate_by_hour.plot(kind='bar', color='red')
plt.title("Procent frauda in fuctie de ore")
plt.xlabel("Ora zilei")
plt.ylabel("Procent tranzactii frauduloase ")
plt.xticks(rotation=0)
plt.show()
```

Am analizat distributia tranzactiilor atat frauduloase cat si normale in functie de ore observam ca cea mai mare rata de frauda este noapte mai ales in intervalul orelor 2-7 dimineata

```python
plt.figure(figsize=(14,7))
sns.scatterplot(
    data=df.sample(20000, random_state=42),
    x='HourOfDay',
    y='Amount',
    hue='Class',
    alpha=0.5,
    palette=['cyan','black']
)

plt.title('Amount vs HourOfDay')
plt.xlabel('Ora zilei')
plt.ylabel('Suma tranzacției')
plt.show()
```

# Aici observam ca la sume mari fraudele sunt aproape inexistente
# Iar pe ore nu exista un inerval sigur, fraudatori actioneaza in mai multe momente ale zilei doar la sume mici

```python
# Creem binu-uri pentru Amount ( adica impatim sumele in intervale de valori  )
df['AmountBin']= pd.cut(df['Amount'], bins=[0,10,50,100,500,2000,30000])
fraud_rate_by_amount = df.groupby('AmountBin')['Class'].mean() * 100
```

```python
plt.figure(figsize=(14,6))
fraud_rate_by_amount.plot(kind='bar', color='darkred')

plt.title("Procent fraudă în funcție de intervalele Amount")
plt.xlabel("Interval Amount")
plt.ylabel("Procent tranzacții frauduloase")
plt.xticks(rotation=45)
plt.show()
```

„Graficul prezintă procentul de tranzacții frauduloase în funcție de intervalele de valori ale sumei tranzacției. Se observă că rata de fraudă este mai ridicată în intervalele de sume mici și medii, în timp ce tranzacțiile cu valori mari prezintă o rată de fraudă mai scăzută. Această distribuție sugerează că fraudele sunt mai frecvente la sume reduse și evidențiază necesitatea utilizării unor modele capabile să detecteze fraude chiar și la valori mici ale tranzacției.”

```python
fraud_matri=df.pivot_table(index='HourOfDay', columns='AmountBin', values='Class', aggfunc='mean')
# Am creat o matrice care calculeaza media de frauda.
#Aceasta ne permite sa indentificam pattern-uri de comportament fraudulor in functi de ora si suma
```

```python
plt.figure(figsize=(14,7))
sns.heatmap(
    fraud_matri * 100,
    cmap='Reds',
    annot=False,
    cbar_kws={'label': 'Procent fraudă'}
)

plt.title('Procent fraudă în funcție de ora zilei și intervalele de Amount')
plt.xlabel('Interval Amount')
plt.ylabel('Ora zilei')
plt.tight_layout()
plt.show()
```

Concluzie Generala EDA :

„Analiza datelor arată un dezechilibru semnificativ între tranzacțiile frauduloase și cele normale, fraudele reprezentând un procent foarte redus din total. Distribuția tranzacțiilor pe parcursul zilei indică faptul că fraudele apar la toate orele, cu variații minore, ceea ce sugerează că ora tranzacției poate avea relevanță, dar nu este suficientă pentru identificarea fraudelor în mod izolat.

Analiza valorii tranzacțiilor evidențiază faptul că fraudele sunt mai frecvente la sume mici și medii, însă nu există un prag clar al sumei care să permită identificarea directă a unei tranzacții frauduloase. De asemenea, analiza combinată a orei și a sumei nu relevă pattern-uri evidente.

În concluzie, rezultatele EDA sugerează că detectarea fraudelor necesită utilizarea unor modele de învățare automată și a unor tehnici adecvate pentru gestionarea dezechilibrului dintre clase.”

3 . Preprocesare

```python
df_copy = pd.read_csv("creditcard.csv")
# creem o copie a data framului pentru a face preprocesare si antrenarea modelelor
```

```python
df_copy.isna().sum().sum()
```

```python
df_copy = df_copy.dropna(subset=['Class']).copy()
df_copy['Class'] = df_copy['Class'].astype(int)
```

```python
X = df_copy.drop(columns='Class')
y = df_copy['Class']
```

```python
#### Split train / test
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2,stratify=y,random_state=42)
```

```python
### Scalare
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled  = scaler.transform(X_test)
```

```python
# Modele Fara Smote
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
```

```python
models_no_smote = {
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
    "XGBoost": XGBClassifier(
        eval_metric='logloss',
        use_label_encoder=False,
        random_state=42
    )
}
```

```python
from sklearn.metrics import classification_report, confusion_matrix

for name, model in models_no_smote.items():
    model.fit(X_train_scaled, y_train)
    y_pred = model.predict(X_test_scaled)

    print(f"\n{name} – WITHOUT SMOTE")
    print(confusion_matrix(y_test, y_pred))
    print(classification_report(y_test, y_pred))
```

VARIANTA SMOTE


```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)

X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)
```

```python
y_train_smote.value_counts()
```

```python
## SCALARE SMOTE
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_smote_scaled = scaler.fit_transform(X_train_smote)
X_test_scaled = scaler.transform(X_test)
```

```python
models_smote = {
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
    "XGBoost": XGBClassifier(
        eval_metric='logloss',
        random_state=42
    )
}
```

```python
for name, model in models_smote.items():
    model.fit(X_train_smote_scaled, y_train_smote)
    y_pred = model.predict(X_test_scaled)

    print(f"\n{name} – WITH SMOTE")
    print(confusion_matrix(y_test, y_pred))
    print(classification_report(y_test, y_pred))
```

```python
results_df = pd.DataFrame({
    "Model": [
        "Logistic Regression", "Random Forest", "XGBoost",
        "Logistic Regression", "Random Forest", "XGBoost"
    ],
    "SMOTE": [
        "Fără SMOTE", "Fără SMOTE", "Fără SMOTE",
        "Cu SMOTE", "Cu SMOTE", "Cu SMOTE"
    ],
    "Precision": [0.83, 0.94, 0.87, 0.13, 0.84, 0.81],
    "Recall":    [0.63, 0.82, 0.80, 0.90, 0.83, 0.86],
    "F1":        [0.72, 0.87, 0.83, 0.23, 0.83, 0.83]
})
```

```python
comparative_table = results_df.pivot_table(
    index="Model",
    columns="SMOTE",
    values=["Precision", "Recall", "F1"]
)

comparative_table
```

Tabelul comparativ evidențiază diferențele de performanță ale modelelor analizate, cu și fără aplicarea SMOTE.
Se observă o creștere semnificativă a recall-ului pentru clasa de fraudă în cazul Logistic Regression după aplicarea SMOTE, însoțită de o scădere a precision-ului.
Modelele ensemble (Random Forest și XGBoost) prezintă performanțe mai stabile, XGBoost oferind cel mai bun compromis între precision și recall în varianta cu SMOTE
